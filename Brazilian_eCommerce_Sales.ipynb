## Merge Data

### Merge Orders and Items Datasets

```python
# merge Orders and Order Items
df_merged_orders_items = pd.merge(df_orders, 
                                  df_items, 
                                  on='order_id')
df_merged_orders_items.head()
```

### Merge Products Dataset into Joined Data

```python
# merge Orders/Order Items and Products
df_merged_orders_products = pd.merge(df_merged_orders_items, 
                                     df_products, 
                                     on='product_id')
df_merged_orders_products.head()
```

### Merge Customers Dataset into Joined Data

```python
# merge Orders/Order Items/Products and Customers
df_merged_orders_customers = pd.merge(df_merged_orders_products, 
                                      df_customers, 
                                      on='customer_id')
df_merged_orders_customers.head()
```

### Merge Sellers Dataset into Joined Data

```python
# merge Orders/Order Items/Products/Customers and Sellers
df_merged_orders_sellers = pd.merge(df_merged_orders_customers, 
                                    df_sellers, 
                                    on='seller_id')
df_merged_orders_sellers.head()
```

### Merge Payments Dataset into Joined Data

```python
# merge Orders/Order Items/Products/Customers/Sellers and Payments
df_merged_orders_payments = pd.merge(df_merged_orders_sellers, 
                                     df_payments, 
                                     on='order_id', 
                                     how='left')
df_merged_orders_payments.head()
```

### Merge Reviews Dataset into Joined Data

```python
# merge Orders/Order Items/Products/Customers/Sellers/Payments and Reviews
df_merged_orders_reviews = pd.merge(df_merged_orders_payments, 
                                    df_reviews, 
                                    on='order_id', 
                                    how='left')
df_merged_orders_reviews.head()
```

### Merge Product Category Translation into Joined Data

```python
# merge Orders/Order Items/Products/Customers/Sellers/Payments/Reviews and Product Category Translation
df_merged_orders_translated = pd.merge(df_merged_orders_reviews, 
                                       df_prod_categories, 
                                       on='product_category_name', 
                                       how='left')
df_merged_orders_translated.head()
```

### Export Joined Data into CSV

```python
# save merged dataset into CSV
df_merged_orders_translated.to_csv('data/super_dataset.csv')
```
